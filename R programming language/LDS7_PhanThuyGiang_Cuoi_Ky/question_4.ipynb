**Logistic Regression**

```r
# tao train/test
set.seed(42)
trainingRowIndex <- sample(1:nrow(m.data), 0.8*nrow(m.data), replace = FALSE)
trainingRowIndex
```

```r
#train data
trainData <- m.data[trainingRowIndex, ]
nrow(trainData)
```

```r
#test
testData <- m.data[-trainingRowIndex, ]
nrow(testData)
```

```r
mylogit <- glm(Class~., data = trainData, family = "binomial")
summary(mylogit)
```

```r
pred = predict(mylogit,newdata = testData,type = "response")
pred_value <- ifelse(pred > 0.5, 1, 0)
print("Testdata Class vs predict (10 rows:)")
result <- data.frame(testClass = testData$Class[30:40], pred_value[30:40])
result
```

```r
# SOLUTION 1
accuracy <- table(pred_value, testData[,"Class"])
accuracy = sum(diag(accuracy))/sum(accuracy)
print(paste("Accuracy s1:", accuracy))

# SOLUTION 2
misClasificError <- mean(pred_value != testData$Class)
print(paste('Accuracy s2: ',1-misClasificError))
```

**Decision Tree**

```r
set.seed(42)
trainingRowIndex=sample(1:nrow(m.data), 0.8*nrow(m.data))
print("Select training row indexes:")
print(trainingRowIndex)
trainingData=m.data[trainingRowIndex,]
testData=m.data[-trainingRowIndex,]
print("Rows of training data and test data:")
print(nrow(trainingData))
print(nrow(testData))
```

```r
install.packages("rpart.plot")
library(rpart.plot)
```

```r
mushroom.tree=rpart(Class~., data=trainingData, method="class")
mushroom.tree
```

```r
prp(mushroom.tree, type=2, extra="auto", nn=TRUE, branch=1, varlen=0, yesno=2)
```

```r
y_pred=predict(mushroom.tree, testData, type="class")
```

```r
result=table(y_pred, testData$Class)
result
```

```r
acc=sum(diag(result))/sum(result)
acc
# Độ chính xác của Model rất cao
```
