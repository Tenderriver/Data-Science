```r
sum_even.function=function(start, end){
  sum_even=0
  for (i in start:end){
    if (i%%2==0){
      sum_even=sum_even+i
    }
  }
  print(sum_even)
}
```

```r
sum_even.function=function(start, end){
  sum_even=0
  for (i in start:end){
    if (i%%2==0){
      sum_even=sum_even+i
    }
  }
  return(sum_even)
}
```

```r
print(paste("Sum of even numbers is", sum_even.function(1,10)))
```

```r
sum=sum_even.function(1,10)
```

```r
print(sum)
```

```r
print(paste("Sum of even numbers is", sum_even.function(start=1, end=10)))
```

```r
print(paste("Sum of even numbers is", sum_even.function(end=10, start=1)))
```
