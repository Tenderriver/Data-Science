**Exercise 1: Thông tin cá nhân**

```r
name = readline(prompt="Enter name: ")
year = as.integer(readline(prompt="Enter year of birth: "))
height = readline(prompt="Enter heigh(m): ")
height = as.double(height)
weight = readline(prompt="Enter weight: ")
weight = as.double(weight)
```

```r
bmi = weight/(height^2)
print(paste("Name:", name, ", Year of Birth:", year))
print(paste("Height:", round(height,2), "(m), Weigth:", round(weight,2), "(kg), BMI: ", round(bmi,2)))
```

**Exercise 2: Tính tiền hóa đơn ăn uống**

```r
sum_of_food_and_drink = as.integer(readline(prompt="Total of foods and drinks: "))
tax = as.integer(readline(prompt="Tax (10-20 %): "))
tip = as.integer(readline(prompt="Tip (5-10 %)/:"))
```

```r
tax.m = sum_of_food_and_drink * tax/100
tip.m = sum_of_food_and_drink * tip/100
total = sum_of_food_and_drink + tax.m + tip.m
```

```r
print(paste("Total of foods and drinks: ",sum_of_food_and_drink, "VND"))
print(paste("Tax: ", tax.m, "VND"))
print(paste("Tip: ", tip.m, "VND"))
print(paste("You need to pay (total): ",format(total,scientific = FALSE), "VND"))
```

**Exercise 3: Tính tiền lãi gửi tiết kiệm**

```r
lai_suat_nam = as.numeric(readline(prompt = "Input interest rate/year (%): "))
so_tien = as.numeric(readline(prompt = "Input amount of money:"))
so_ngay_gui = as.integer(readline(prompt = "Input number of days:"))

lai_suat_ngay = (lai_suat_nam/365)/100
tien_lai = so_tien * lai_suat_ngay * so_ngay_gui
tien_gui_va_lai = so_tien + tien_lai
```

```r
result = paste("Interest rate/day: ", lai_suat_ngay,
"\nAmount of money: ", so_tien, " after ", so_ngay_gui,
" days => Saving interest money: ", round(tien_lai,2),
"\nSum:", format(tien_gui_va_lai, big.mark=".", decimal.mark = ",", nsmall = 2),sep="")

cat(result)
```

**Exercise 4: Tính bán kính và chu vi hình tròn**

```r
S = readline(prompt = "Input S of cricle:")
S = as.double(S)

PI = 3.14
R = sqrt(S/PI)
P = 2 * PI * R
```

```r
print(paste("R of circle:", round(R,2)))
print(paste("P of circle:", round(P,2)))
```

**Exercise 5: Làm việc với chuỗi**

```r
str1 = readline(prompt = "Input s1:")
str2 = readline(prompt = "Input s2:")
str3 = readline(prompt = "Input s3:")
index = as.integer(readline(prompt = "Input index:"))
str_find = readline(prompt = "Find_string:")
str_replace = readline(prompt = "Replace_string:")
```

```r
str4 = substring(str1,index)
str5 = substring(str1, 1, index)
# thay the 1 chuoi voi 1 chuoi dau tien thoa
str_new_1 =sub(str_find, str_replace, str1)
#thay the 1 chuoi cho tat ca cac chuoi thoa
str_new_2 =gsub(str_find, str_replace, str1)
```

```r
print(paste("str4 (index ->):", str4))
print(paste("str5 (->index):", str5))
print(paste("Lenght of str1:", nchar(str1),
", Length of str2:", nchar(str2),
", Length of str3:", nchar(str3)))
print(paste("After replace:", str_new_1))
print(paste("After replace:", str_new_2))
```

**Exercise 6: Khoảng cách giữa 2 ngày**

```r
date = readline(prompt = "Input date: dd-mm-YYYY:")
date = as.Date(date, format("%d-%m-%Y"))
```

```r
now_date = Sys.Date()
distance = now_date - date
```

```r
print(paste("Input date:", date))
print(paste("Now:", now_date))
print(paste("Distance between", now_date, "and", date, "is", distance))
```
