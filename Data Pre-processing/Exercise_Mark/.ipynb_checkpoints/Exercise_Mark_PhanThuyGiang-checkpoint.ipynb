### Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo State và Year 

* Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('us-state-college-pricing-2017.csv',skiprows=2)
```

```python
df.head()
```

```python
df.info()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
df.replace({'\$':'',',':''}, regex = True,inplace=True)
```

```python
df.head()
```

```python
column=list(df.iloc[:,1:].columns)
column
```

```python
df[column]=df[column].astype(int)
```

```python
df.info()
```

```python
df1=df.iloc[:,:15]
df1.head()
```

```python
df1_new = pd.melt(frame=df1, id_vars='State', var_name='College_Year', value_name='In-District 2 Year Tuition Fee')
df1_new
```

```python
df2=df.drop(columns=column[:14])
df2 = df2.set_axis(df1.columns, axis=1)
df2.head()
```

```python
df2_new = pd.melt(frame=df2, id_vars='State', var_name='College_Year', value_name='In-State 2 Year Tuition Fee')
df2_new
```

```python
df_new = pd.merge(left=df1_new, right=df2_new, on=['State','College_Year'])
df_new.head()
```

```python
df1_line=pd.DataFrame(df1.mean().round(2),columns=['mean'])
df1_line.reset_index(inplace=True)
df1_line
```

```python
df2_line=pd.DataFrame(df2.mean().round(2),columns=['mean'])
df2_line.reset_index(inplace=True)
df2_line
```

```python
plt.figure(figsize=(10, 7))

sns.lineplot(
    data=df1_line,
    x="index", y="mean",
    markers='x', dashes=False)

plt.xlabel('College_Year')
plt.ylabel('Avg_In-District 2 Year Tuition Fee')
for x, y in zip(df1_line['index'], df1_line['mean']):
    plt.text(x = x, y = y-50, s = '$'+'{:.2f}'.format(y), color = 'purple')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 7))

sns.lineplot(
    data=df2_line,
    x="index", y="mean",
    markers='x', dashes=False)

plt.xlabel('College_Year')
plt.ylabel('Avg_In-State 2 Year Tuition Fee')
for x, y in zip(df2_line['index'], df2_line['mean']):
    plt.text(x = x, y = y-50, s = '$'+'{:.2f}'.format(y), color = 'purple')
plt.xticks(rotation=45)
plt.show()
```

```python
df1_pivot = df1_new.pivot(index='State', columns='College_Year', values='In-District 2 Year Tuition Fee')
df1_pivot.reset_index(inplace=True)
df1_pivot.index.name = None
df1_pivot.columns.name = None
df1_pivot
```

```python
df2_pivot = df2_new.pivot(index='State', columns='College_Year', values='In-State 2 Year Tuition Fee')
df2_pivot.reset_index(inplace=True)
df2_pivot.index.name = None
df2_pivot.columns.name = None
df2_pivot
```
