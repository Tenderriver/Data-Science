```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url="https://www.hubertiming.com/results/2018MLK"     #openlink
html=urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
title = soup.title
print(title)
print(title.text)
```

```python
links = soup.find_all('a', href=True)
for link in links:
    print(link.get("href"))
```

```python
data = []
allrows = soup.find_all("tr")
for row in allrows:
    row_list = row.find_all("td")
    dataRow = []
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)
data  = data[5:]
print(data[0:5])
```

```python
df = pd.DataFrame(data)
df.head()

```

```python
header_list = []
col_headers = soup.find_all('th')
for col in col_headers:
    header_list.append(col.text)
print(header_list)
```

```python
df.columns = header_list
df.head()
```

```python
df['Name'] = df['Name'].str.replace('\r\n\r\n','')
df['Gender Place'] = df['Gender Place'].str.replace('\r\n\r\n','')
df['Age Group Place'] = df['Age Group Place'].str.replace('\r\n\r\n','')
```

```python
df
```

```python
df1=df.copy()
df1[['h','m','s']]=df1['Chip Time'].str.split(':',expand=True)
df1.info()
```

```python
df1
```

```python
df1['s']=df1['s'].fillna(value='0')
df1['s']
```

```python
df1['h']=df1['h'].astype(int)
df1['m']=df1['m'].astype(int)
df1['s']=df1['s'].astype(int)
df1.info()
```

```python
df1['timestamp']=df1['h']*3600+df1['m']*60+df1['s']
df1.head()
```

```python
df1['test'] = pd.to_datetime(df1['timestamp'],unit='s').dt.strftime('%H:%M:%S')
df1
```
