# Monthly champagne sales millions – Time series Analysis

```r
install.packages("xlsx")
library("xlsx")
```

```r
#https://docs.google.com/spreadsheets/d/1QguzL5grPwF4jpAUzbHiRtmfDEn5fnQZ/edit?usp=sharing&ouid=100667830146032102250&rtpof=true&sd=true
system('gdown --id 1QguzL5grPwF4jpAUzbHiRtmfDEn5fnQZ')
```

```r
df <- read.xlsx("champagne_new.xlsx", sheetName ="monthly_champagne")
head(df)
```

```r
nrow(df)
```

```r
ncol(df)
```

```r
str(df)
```

```r
summary(df)
```

```r
ts_df = ts(df['monthly_champange'], frequency = 12, start = c(1964, 1), end = c(1972, 9))
```

```r
ts_df
```

```r
plot.ts(ts_df, main = 'Champaign sale from 01/1964 to 09/1972',
        xlab= "year", ylab = 'month', col = 'blue')
options(repr.plot.width =12, repr.plot.height = 8)
```

```r
comp = decompose(ts_df)
comp$x
comp$trend
comp$seasonal
comp$random 
plot(comp)
```

**Nhận xét**
1. Trend: Tăng dần qua các năm
2. Có yếu tố theo mùa, bán nhiều nhất vào 3 tháng cuối năm, tháng 10- tháng 12

```r
forecast = HoltWinters(ts_df, beta = F, gamma = F)
plot(forecast)
# Sum of squared errors
forecast$SSE
#RMSE
(forecast$SSE/length(ts_df))^0.5
```

```r
install.packages('forecast')
library('forecast')
```

```r
forecast.new = forecast:::forecast.HoltWinters(forecast, h=9) 
# predict 9 month
forecast.new
```

```r
forecast:::plot.forecast(forecast.new)
```
