1. Liệt kê các sản phẩm của nước Nhật theo mẫu sau, sắp tăng theo city

```python
%%sql
SELECT s.City, s.CompanyName, p.ProductName, p.UnitPrice
FROM products p LEFT JOIN suppliers s 
ON p.SupplierId=s.Id
WHERE s.Country="Japan"
ORDER BY s.City
```

2. Thống kê theo quốc gia, cần tính đơn giá trung bình, đếm số sản phẩm, đếm số thành phố.  Sắp tăng theo count_product

```python
%%sql
SELECT s.Country, CONCAT('$',ROUND(AVG(p.UnitPrice),2)) avg_unitprive, COUNT(p.ProductName) count_product, COUNT(DISTINCT s.City) count_city
FROM  suppliers s LEFT JOIN  products p
ON p.SupplierId=s.Id
WHERE p.IsDiscontinued=0
GROUP BY s.Country
ORDER BY count_product
```

3. Liệt kê các đơn đặt hàng đặt trong tháng 3/2014 theo mẫu sau, sắp giảm dần theo  orderdate và totalAmount 

```python
%%sql
SELECT CONCAT(c.FirstName,' ',c.LastName) Customer_name, o.OrderNumber, CONCAT(DAY(o.OrderDate),'/',MONTH(o.OrderDate),'/',YEAR(o.OrderDate)) order_date, CONCAT('$',o.TotalAmount) total_amount
FROM orders o JOIN customers c ON o.CustomerId =c.Id
WHERE MONTH(o.OrderDate)=3 and YEAR(o.OrderDate)=2014
ORDER BY o.OrderDate DESC, o.TotalAmount DESC
```

4. Cho biết theo mỗi năm với 5 sản phẩm có tổng thành tiền lớn nhất

```python
%%sql
SELECT sale_year, product_name, sum_amount
FROM
(SELECT a.sale_year, a.product_name, a.sum_amount, RANK() OVER(PARTITION BY a.sale_year ORDER BY a.sum_amount DESC) hang
FROM
(SELECT YEAR(o1.OrderDate) sale_year, p.ProductName product_name, SUM(o.UnitPrice*o.Quantity) sum_amount
FROM orderitems o LEFT JOIN orders o1 ON o.OrderId=o1.Id
LEFT JOIN products p ON p.Id=o.ProductId
GROUP BY sale_year, product_name) a) b
WHERE hang<6
```

5. Liệt kê các đơn đặt hàng theo mẫu sau và có ngày đặt hàng là ngày chủ nhật của tháng 3/2014, sắp giảm theo orderdate

```python
%%sql
SELECT o.OrderNumber, o.OrderDate, CONCAT(c.FirstName,' ',c.LastName) Customer_name, CONCAT('$',o.TotalAmount) total_amount
FROM orders o JOIN customers c ON o.CustomerId =c.Id
WHERE MONTH(o.OrderDate)=3 and YEAR(o.OrderDate)=2014 and WEEKDAY(o.OrderDate)=6
ORDER BY o.OrderDate DESC
```

6. Liệt kê các customer theo mẫu sau và đếm các đơn đặt hàng của các customer, chỉ in các customer nào có đếm>15, sắp giảm theo đếm


```python
%%sql
SELECT CONCAT(c.FirstName,' ',c.LastName) Customer_name, c.Country, c.City, count(o.CustomerId) count_order
FROM orders o JOIN customers c ON o.CustomerId =c.Id
GROUP BY o.CustomerId
HAVING count_order >15
ORDER BY count_order DESC
```

7. Cho biết 2 quý nào có tổng thành tiền bán cao nhất

```python
%%sql
SELECT YEAR(`OrderDate`) year, Floor((MONTH(`OrderDate`)+2)/3) quarter, SUM(`TotalAmount`) sum_totalamount
FROM orders
GROUP BY year, quarter
ORDER BY sum_totalamount DESC
LIMIT 2
```

8. Liệt kê tất cả khách hàng và đếm số đơn đặt hàng, sắp tăng theo count_orders

```python
%%sql
SELECT CONCAT(c.FirstName,' ',c.LastName) Customer_name, COUNT(o.CustomerId) count_order
FROM  customers c LEFT JOIN orders o  ON c.Id=o.CustomerId
GROUP BY Customer_name
ORDER BY count_order
```

9. Liệt kê theo mẫu sau gồm 10 đơn đặt hàng có đặt product tên là Chai, sắp giảm theo  totalamount

```python
%%sql
SELECT o1.Id,o1.OrderDate,o1.CustomerId,o1.TotalAmount
FROM orderitems o LEFT JOIN orders o1 ON o.OrderId=o1.Id
LEFT JOIN products p ON p.Id=o.ProductId
WHERE p.ProductName LIKE '%Chai'
ORDER BY o1.TotalAmount DESC
LIMIT 10
```

10. Cho câu truy vấn như sau
∙ Hãy giải thích ý nghĩa của câu truy vấn trên 
∙ Thay thế NULL của cột sum_quantity thành giá trị 0 

Câu truy vấn dùng để tính số lượng các sản phẩm vẫn còn được sản xuất, sắp theo thứ tự số lượng giảm dần

```python
%%sql
SELECT p.ProductName, COALESCE(SUM(i.Quantity),0) sum_quantity
FROM products p LEFT JOIN orderitems i ON p.Id=i.ProductId
WHERE p.IsDiscontinued=0
GROUP BY p.ProductName
ORDER BY sum_quantity
```
