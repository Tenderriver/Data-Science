### Hãy viết chương trình in bàn cờ 8x8.
image.png

```r
chessboard.function = function(n){
  for (i in 1:n){
    message(replicate(n,'-----'))
    for (j in 1:n){
      if ((i + j)%%2 ==0){
        color = 'W'
      } else {
        color = 'B'
      }
    cat(paste('|', color, '|'))
    }
    message()
  }
  message(replicate(n,'-----'))
}

chessboard.function(8)
```

## Câu 2: Vector (0.5đ)
### Yêu cầu : Viết chương trình xử lý vector như sau
1. Chương trình sẽ tạo ra 2 vector : vec1 và vec2 như sau :
    -  Người dùng nhập số lượng phần tử cho vec1, vec2 lần lượt là n, m
    -  Hệ thống tạo ra n phần tử số nguyên ngẫu nhiên cho vec1, m phần tử số nguyên ngẫu nhiên cho vec2
2. Cho biết những phần tử chỉ có trong vec1 mà không có trong vec2
3. Cho biết những phần tử xuất hiện trong cả vec1 và vec2
4. Tạo 1 vector mới là vec12 chứa tất cả các phần tử của vec1 và vec2. Thống kê số lần xuất hiện của từng phần tử. Phần tử nào xuất hiện nhiều lần nhất ?
5. Tạo 1 vector mới là vec_unique chỉ chứa những phần tử duy nhất từ vec12
### Ví dụ :
- Người dùng nhập n = 3
- Người dùng nhập m= 2
- Hệ thống tạo ra: vec1 = c(23, 18, 7), vec2 = c(15, 18)
- Phần tử xuất hiện trong cả vec1 và vec2 là: 18
- vec12 = c(23, 18, 7, 15, 18)
#### Thống kê:
- 23: xuất hiện 1 lần
- 18: xuất hiện 2 lần
- 7: xuất hiện 1 lần
- 15: xuất hiện 1 lần
- Phần tử xuất hiện nhiều nhất là 18 với 2 lần
#### vec_unique = c(23, 18, 7, 15)

```r
n = as.integer(readline(prompt = 'Input n: '))
m = as.integer(readline(prompt = 'Input m: '))

vec1 = sample(1:100, n)
print(paste('Vector vec1:', toString(vec1)))
vec2 = sample(1:100, m)
print(paste('Vector vec2:', toString(vec2)))

vec1_not_vec2 = setdiff(vec1, vec2)
print(paste('Phan tu chi xuat hien trong vec1, khong co trong vec2:',  toString(vec1_not_vec2)))

vec1_and_vec2 = intersect(vec1, vec2)
print(paste('Phan tu xuat hien trong ca vec1 va vec2:',  toString(vec1_and_vec2)))

vec12 = c(vec1, vec2)

for (i in 1:length(unique(vec12))){
  number = names(table(vec12)[i])
  count = as.numeric(table(vec12)[i])
  print(paste(number, ': xuat hien', count, 'lan'))
}

max_name = names(table(vec12)[table(vec12) == max(table(vec12))])
max_value = max(table(vec12))

cat('Phan tu xuat hien nhieu nhat la:\n', (paste(max_name, 'voi', max_value, 'lan', '\n')))

vec_unique = unique(vec12)
print(paste('Vector vec_unique:', toString(vec_unique)))
```
