```
# Nhập vào ngày sinh của mình theo định dạng: 2021-mm-dd.
date_birth <- readline(prompt = 'Input your date of birth YYYY-MM-DD: ')
date_birth <- as.Date(date_birth, format='%Y-%m-%d')
date_birth
```

```
# Lấy ngày hiện hành
Sys.Date()
```

```
# Hãy in ra một trong 3 nội dung sau tùy kết quả (so với ngày hiện hành):
# Nếu ngày sinh nhật của bạn chưa tới thì "The next xx days will be your birthday."
# Nếu ngày sinh nhật của bạn dã qua thì "Your birthday has passed xx days."
# Nếu hôm nay là sinh nhật của bạn thì "Happy birthday to you!"
```

```
today = as.Date(Sys.Date(), format="%d-%m")
```

```
dif = as.numeric(difftime(date_birth, today, units = "days"))

if (dif == 0){
  print("Happy birthday")
} else if (dif >0){
  print(paste("Your birthday has passed ", as.character(dif) , " days."))
} else if (dif <0){
  print(paste("The next ", as.character(-dif) , " days will be your birthday."))
}
```

## Bài tập số 3:
### Viết chương trình cho phép người dùng:
- Nhập vào một số n với điều kiện sau: 10<=n<=20. Nếu số người dùng nhập không thỏa điều kiện thì thông báo cho người dùng biết và yêu cầu nhập lại. Nêu thỏa điều kiện thì qua 2.
- Tạo một dãy số v1 có n số mà mỗi số là một số ngẫu nhiên từ 1 đến 10. Ví dụ: v1: 5, 5, 5, 9, 9, 8, 5, 9, 9, 10, 7, 7
- Viết một function tạo dãy số mới v2 từ dãy số v1 nói trên theo điều kiện sau: phần tử đầu tiên và cuối cùng của v2 bằng v1, các phần tử khác thì v2[i] = v1[i] + v1[i-1] + v1[i+1] nếu cả v1[i-1] và v1[i+1] cùng chẵn hoặc cùng lẻ, ngược lại v2[i] = v1[i]. Ví dụ: v2: 5, 15, 19, 23, 9, 22, 5, 23, 9, 26, 7, 7
- Gọi function vừa viết và in kết quả v2.
- Viết function Cho biết trong một dãy số có số chính phương nào hay không?
- Gọi function vừa viết với input là v2 và in kết quả.

```
repeat {
 n = as.numeric(readline(prompt = 'Input n, with 10<=n <=20: '))
 if (n<10 | n>20){
   message("n must satisfies the condition: 10<=n <=20. Please re-input n")
 } else{
   break
 }
}
print(n)
```

```
# Tạo một dãy số v1 có n số mà mỗi số là một số ngẫu nhiên từ 1 đến 10.
# Ví dụ: v1: 5, 5, 5, 9, 9, 8, 5, 9, 9, 10, 7, 7

v1 = sample(1 : 10, size = 10, replace = T)
v1
```

```
# Viết một function tạo dãy số mới v2 từ dãy số v1 nói trên theo điều kiện sau: 
# phần tử đầu tiên và cuối cùng của v2 bằng v1, 
# các phần tử khác thì v2[i] = v1[i] + v1[i-1] + v1[i+1] nếu cả v1[i-1] và v1[i+1] cùng chẵn hoặc cùng lẻ, 
# ngược lại v2[i] = v1[i]. 
# Ví dụ: v2: 5, 15, 19, 23, 9, 22, 5, 23, 9, 26, 7, 7
```

```
odd_list.function <- function(v1){
  v2 <- vector(mode = "list")
  for (i in 1: length(v1)){
    if (i == 1 | i == length(v1)){
      v2 <- append(v2, v1[i])
    } else if ((v1[i-1]+ v1[i+1])%% 2 ==0){
      v2 <- append(v2, (v1[i] + v1[i-1] + v1[i+1]))
    } else {
      v2 <- append( v2, v1[i])
    }
  }
  return (v2)
}
```

```
odd_list.function(v1)
```

```
# Viết function Cho biết trong một dãy số có số chính phương nào hay không?
check_square <- function(v1){
  for ( i in v1){
    if (sqrt(i)*sqrt(i) == i){
      return (print("Yes, there is at least a square number in the list"))
    }
  }
  print('No, there is no square number in the list')
}
```

```
check_square(v1)
```
