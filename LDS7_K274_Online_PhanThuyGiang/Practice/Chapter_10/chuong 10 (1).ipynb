```r
system('gdown --id 1A4KBbxvqfvck5Gu4uC8CIl_TJLc3E_pO')
rainfall = scan('rain_fall.txt')
ts.rain_fall = ts(rainfall, frequency = 12, start = c(2014, 1))
```

```r
ts.rain_fall
```

```r
plot.ts(ts.rain_fall, main = 'Ha Noi rainfall from 01/2014 to 12/2018',
        xlab = 'mm', ylab = 'month', col = 'blue')
```

```r
plot.ts(ts.rain_fall, main = 'Ha Noi rainfall from 01/2014 to 12/2018',
        xlab = 'mm', ylab = 'month', col = 'blue')
options(repr.plot.width =12, repr.plot.height = 8)
```

```r
comp = decompose(ts.rain_fall)
comp$X
comp$trend # trend # không có trend rõ ràng
comp$seasonal # có yếu tố theo mùa, tháng 5-9 mua nhiều, cao nhất tháng 8
comp$random # giá trị thực - giá trị trung bình
plot(comp)
```

```r
# simple exponential smoothing(SES)
forecast.rainfall = HoltWinters(ts.rain_fall, beta = F, gamma = F)
# Holt-winters exponential smoothing without trend and without seasonal compoment

plot(forecast.rainfall)
# Sum of squared errors
forecast.rainfall$SSE
#RMSE
(forecast.rainfall$SSE/length(ts.rain_fall))^0.5
```

```r
install.packages('forecast')
library('forecast') # to predict future
```

```r
forecast.rainfall.new = forecast:::forecast.HoltWinters(forecast.rainfall, h=5) 
# predict 5 month -> mean, khoang tin cay trong do tin cay 80%, 95%
forecast.rainfall.new
```

```r
forecast:::plot.forecast(forecast.rainfall.new)
#plot(forecast.rainfall.new)
```

```r
HES.forecast = HoltWinters(log(ts.rain_fall)) # nên sử dụng log để giảm độ lớn dữ liệu, , gamma = T???
HES.forecast
# alpha: the contribution of past values to predicted values
# beta: trend
# gamma: seasonal, thấp vì nó đang tính theo tháng trước???
# Coeff có 12 giá trị vì có 12 tháng

#rmse
(HES.forecast$SSE/length(ts.rain_fall))^0.5
HES.forecast.new = forecast:::forecast.HoltWinters(HES.forecast, h = 5) # return log value of rainfall
y = data.frame(HES.forecast.new)
y$Forecast = exp(y$Point.Forecast)
y
```

# Distribution

```r
# Normal Distribution
x <- seq(-10, 10, by = 0.1)
y <-dnorm(x, mean = 0, sd =1)
plot(x,y)
```

```r
y[x ==-1]
```

```r
x <-seq(-10, 10, by =1)
y <-pnorm(x, mean = 0, sd =1)
plot(x, y)
```

```r
y[x ==2]
```

```r
x <- seq(0,1,by = 0.02)
y <- qnorm(x, mean = 0, sd = 1)
plot(x, y)
```

```r
y[x ==0.8]
```

```r
y <- rnorm(100000000) # creat 50 samples
hist(y, main = 'Normal distribution')
```
