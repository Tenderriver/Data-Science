```
#C1
odd_even <- function(x) c("odd"=sum(x%%2), "even"=sum(!x%%2))
odd_even(my_matrix)
```

```
#C2
even_odd <- function(x) setNames(table(x %% 2), c("even", "odd"))
even_odd(my_matrix)
```

```
primes=generate_n_primes(100)
v1=sample(primes, 16)
my_matrix=matrix(v1, nrow=4, ncol=4, byrow=TRUE)
```

```
matrix.function <- function(primes){
  v1=sample(primes, 16)
  my_matrix=matrix(v1, nrow=4, ncol=4, byrow=TRUE)
  return (my_matrix)
}
matrix.function(primes)
```

```
# Create my_array (4, 4, 2): from 2 matrices created by calling function create_matrix (Bai tap so 3)

# Use apply to calculate the min of the rows across all the matrices.

# Use apply to calculate the max of the columns across all the matrices.

# Use apply to calculate the sum of each matrix in my_array
```

```
matrix1=matrix.function(primes)
matrix2=matrix.function(primes)
print(matrix1)
print(matrix2)
```

```
result <- array(c(matrix1,matrix2),,dim = c(4,4,2))
print(result)
```

```
i=1
dims=dim(result)[3]
while(i<=dims){
print(sum(apply(result[,,i], 1, sum)))
i <- i+1
}
```
