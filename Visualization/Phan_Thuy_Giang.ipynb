```python
world.head()
```

```python
world['GDP']=world['GDP_CODE'].str.extract(r'(\d+\.\d+)')
world
```

```python
world['GDP']=world['GDP'].astype(np.float64)
world.info()
```

```python
map_1=folium.Map(location=[0,0],zoom_start=3,tiles='cartodbpositron')
map_1
```

```python
geojson=gpd.read_file('world-countries.json')
folium.Choropleth(
    geo_data=geojson,
    data=world,
    columns=['COUNTRY','GDP'],
    key_on='feature.properties.name',
    fill_color='Greens',
    fill_opacity=0.5).add_to(map_1)
map_1
```
