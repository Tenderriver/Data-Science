```python
#phát sinh bộ dữ liệu thử 
from sklearn.datasets import make_swiss_roll
```

```python
X, y = make_swiss_roll(n_samples = 1000, noise=0.2, random_state = 42)
```

```python
#trực quan hóa bộ dữ liệu
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(1, 1, 1, projection = '3d')

ax.view_init(10, -70)
ax.set_title('Swiss Roll Dataset', fontsize=20)
ax.set_xlabel('X1', fontsize=16)
ax.set_ylabel('X2', fontsize=16)
ax.set_zlabel('X3', fontsize=16)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c = y, cmap=plt.cm.hsv)
plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
#khởi tạo LLE
lle = LocallyLinearEmbedding(n_neighbors = 10, n_components=2, random_state = 42)
```

```python
# chuyển đổi X_D -> Y_d
X_reduced = lle.fit_transform(X)
```

```python
X_reduced
```

```python
X_reduced.shape
```

```python
# trực quan hóa sau khi giảm chiều
plt.figure(figsize=(6,6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.hsv)
plt.title('Swiss Roll Dataset after using LLE', fontsize=20)
plt.xlabel('$Z_1$', fontsize=16)
plt.ylabel('$Z_2$', fontsize=16)
plt.grid(True)
plt.show()
```
