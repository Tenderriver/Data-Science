**Exercise 2: Tạo và tính toán trên matrix**

```r
matrix_B <- matrix(data = c(1:12), nrow = 4,
dimnames = list(c("a", "b", "c", "d"), c("x", "y", "z")))
print("Matrix B:")
print(matrix_B)
tB <- t(matrix_B)
print("Transpose of Matrix B:")
print(tB)
subB<-matrix_B[2:dim(matrix_B)[1],1:3]
print("Sub of Matrix B:")
print(subB)
```

```r
matrix_A <- matrix(data = c(1:9), nrow = 3,
dimnames = list(c("a", "b", "c"), c("x", "y", "z")))
print("Matrix A:")
print(matrix_A)
# 3A
A3 <- matrix_A * 3
print("3*A:")
print(A3)
A_plus_subB <- matrix_A + subB
print("A + sub B:")
print(A_plus_subB)
A_minus_subB <- matrix_A - subB
print("A - sub B:")
print(A_minus_subB)
```

**Exercise 3: Tạo và tính toán trên matrix**

```r
n = as.integer(readline(prompt = "Nhap n: "))
```

```r
A1 <- matrix(runif(n*n, min = 1, max = n*n), n, n)
print("A1")
print(A1)
```

```r
m = as.integer(readline(prompt = "Nhap m: "))
```

```r
A2 <- matrix(runif(n*m, min = 1, max = n*m), n, m)
print("A2")
print(A2)
```

```r
M <- solve(A1, A2)
print("Matrix M")
print(M)
```

```r
print("Row 1")
print(M[1,])

print("Column 1")
print(M[,1])
```

```r
if (n>=3 && m>=2){
print("M[3][2]")
print(M[3,2])
}
```

**Exercise 4: Tạo và tính toán trên array**

```r
matrix_1 <- matrix(c(1:9),nrow = 3, byrow = TRUE)
matrix_2 <- matrix(c(10:18),nrow = 3, byrow = TRUE)
A3 <- array(c(matrix_1, matrix_2), dim = c(3, 3, 2),
dimnames = list(c("r1", "r2", "r3"),
c("c1", "c2", "c3"),
c("matrix1", "matrix2")))
print("Array A3:")
print(A3)
```

```r
print("Tong theo dong cua tat ca matrix")
print(apply(A3, c(1),sum))
print("Tong theo cot cua tat ca matrix")
print(apply(A3, c(2),sum))
print("Tong theo matrix cua array")
print(apply(A3, c(3),sum))
print("Tong dong 1, matrix1")
print(sum(A3[1,,1]))
print("Tich cot 2, matrix2")
print(prod(A3[, 2,2]))
```

**Exercise 5: Tạo và tính toán trên array**

```r
v <- c()
i <-1
while(i<=50){
x = floor(runif(1, min = 5, max = 50))
if (x%%2==0){
v[i] <-x
i <- i + 1
}
}
print(v)
```

```r
A4 <- array(v, dim=c(5, 5, 2))
print("Array A4:")
print(A4)
```

```r
rows <- nrow(A4)
print(rows)
```

```r
cols <- ncol(A4)
print(cols)
```

```r
dims <- dim(A4)[3]
print(dims)
```

```r
i <- 1
j <- 1
k <- 1
tong <-0
while(i<=dims){
j <-1
while(j<=rows){
k <-1
while(k<=cols){
if(A4[j,k,i]%% 2 == 0 & A4[j,k,i]%% 3 == 0 & A4[j,k,i]%% 6 == 0){
print(paste("A4[", j,",", k,",", i,"] =", A4[j,k,i]))
tong <- tong + A4[j,k,i]
}
k <- k+1
}
j <- j+1
}
i <- i+1
}
print(paste("Tong =", tong))
```

**Exercise 7: Tạo và tính toán trên list**

```r
year <- c(2008:2018)
month <- c(1:12)
day <- c(1:31)
# list
year_month_day.list <- list("year" = year, "month" = month, "day" = day)
print("List year_month_day:")
print(year_month_day.list)
```

```r
year_month_day.list$year <- c(2010:2020)
print("List year_month_day after changing:")
print(year_month_day.list)
```

```r
year_month_day.list$year <- year_month_day.list$year[-1]
print("After drop the first year in year:")
print(year_month_day.list)
```

```r
x <- c(1,3,4,7,11,18,29)
cal.list <- list("x*2"= x*2,"x/2"=x/2, "sqrt(x)"=sqrt(x))
print("List calculator:")
print(cal.list)
```

```r
#tra ve cac phan tu 3-5 $sqrt(x)
print("3 phan tu giua")
print(cal.list$`sqrt(x)`[3:5])
```

```r
#tao 1 list moi co 2 list tren
year_month_day_cal.list <- c(year_month_day.list, cal.list)
print("year_month_day_cal.list: ")
print(year_month_day_cal.list)
```

```r
# tao mot list con chua year, x*2 va day tu list moi tao ra tren
year_x2_day <- year_month_day_cal.list[c("year", "x*2", "day")]
print("list year_x2_day:")
print(year_x2_day)
```

```r
# chieu dai cua x*2
print(paste("length of x*2:", length(year_x2_day$`x*2`)))
print(paste("last value of year:",
year_x2_day$year[length(year_x2_day$year)]))
```
