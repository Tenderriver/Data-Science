```r
x=c(46, 37, 35, 53)
labels=c("10A1", "10A2", "10A3", "10A4")
pie(x, labels)
dev.off()
```

```r
pie(x, labels, main="Classes pie chart", col=rainbow(length(x)), clockwise=TRUE)
dev.off()
```

```r
piepercent=round(100*x/sum(x),1)
```

```r
pie(x, labels=paste(piepercent,"%"), main="Classes pie chart", col=rainbow(length(x)), clockwise=TRUE)
legend('topright', labels, cex=0.8, fill=rainbow(length(x)))
dev.off()
```

```r
library(plotrix)
```

```r
pie3D(x, labels=labels, main="Classes Pie Chart",col=rainbow(length(x)), explode=0.05)
dev.off()
```

```r
H=c(5,15,20,5)
M=c("Excellent", "Very Good", "Good", "Medium")
barplot(H, names.arg=M, xlab="Grade", ylab="Students", col="blue", main="Grade of class 10A1", border="red")
dev.off()
```

```r
# Create the input vectors.
colors <- c("green","orange","brown", "red") # for 4 grades
classes <- c("10A1","10A2","10A3")
grades <- c("Excellent","Very Good","Good","Medium")
# Create the matrix of the values.
Values <- matrix(c(5, 6, 7,15, 14, 13, 20, 21, 22, 5, 4, 3),
                 nrow = 4,ncol = 3,byrow = TRUE)
print(Values)
```

```r
barplot(Values, names.arg=classes, xlab="Class", ylab="Students", col=colors, main="Grade of 3 classes")
legend("right", grades, cex=0.9, fill=colors)
dev.off()
```

```r
bears=c(8,54,93,116,137)
```

```r
plot(bears, type='o', col='red', xlab="Year", ylab="Numbers", main="Number of bears chart")
dev.off()
```

```r
plot(bears, type='o', col='red', xlab="Year", ylab="Numbers", main="Number of bears chart", xaxt="n")
axis(1,at=1:5, labels=2017:2021)
dev.off()
```

```r
bears=c(8,54,93,116,137)
dolphins=c(150,77,32,11,6)
whales=c(80,54,100,76,93)
g_range=range(0,bears, dolphins, whales)
plot(bears, type="o", col="red", xlab="Year", ylab="Numbers", main="Numbers of animal", xaxt='n', ylim=g_range)
axis(1,at=1:5, labels=2017: 2021)
lines(whales, type="o", col="green")
legend("top", inset=0.01, c("bear", "dolphins", "whales"), cex=0.8, fill=c("red", "blue", "green"))
dev.off()
```
