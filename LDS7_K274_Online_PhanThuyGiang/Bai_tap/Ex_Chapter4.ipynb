**Exercise 2: Bình phương vector**

```r
n <- as.integer(readline(prompt = "Please input n: "))
```

```r
vector_random = round(runif(n,1,10), 0)
print(vector_random)
```

```r
square = c()
for (value in vector_random){
square <- c(square, value^2)
}
```

**Exercise 5: Đoán số**

```r
random_number = round(runif(1,1,10), 0)
random_number

times <- 1
flag = FALSE
```

```r
while (times<=3){
your_number <- as.integer(readline(prompt = "Your number (1..10): "))
if(your_number == random_number){
message("You win!!!")
flag = TRUE
break
}
else if(your_number > random_number){
message("Your number > Our number")
}
else{
message("Your number < Our number")
}
times <- times + 1
}
if(flag == FALSE){
message("You lose!!!")
}
```
